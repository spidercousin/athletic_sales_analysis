```python
# Import Libraries and Dependencies
import pandas as pd





```

### 1. Combine and Clean the Data
#### Import CSVs

```python
# Read the CSV files into DataFrames.
athletic20 = pd.read_csv('Resources/athletic_sales_2020.csv')
athletic21 = pd.read_csv('Resources/athletic_sales_2021.csv')

sales20 = pd.DataFrame(athletic20)
sales21 = pd.DataFrame(athletic21)
```

```python
# Display the 2020 sales DataFrame
sales20.head()
```

```python
# Display the 2021 sales DataFrame
sales21.head()
```

#### Check the data types of each DataFrame

```python
# Check the 2020 sales data types.
sales20.info()
```

```python
# Check the 2021 sales data types.
sales21.info()
```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.
combined_data = pd.concat([sales20,sales21], axis=0, join="inner").reset_index(drop=True)
combined_data
```

```python
# Check if any values are null.
combined_data.isnull().any()
```

```python
# Check the data type of each column
combined_data.info()
```

```python
# Convert the "invoice_date" to a datetime datatype

combined_data['invoice_date'] = pd.to_datetime(combined_data['invoice_date'])

```

```python
# Confirm that the "invoice_date" data type has been changed.
combined_data.info()
```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number products sold for region, state, and city.
# Rename the sum to "Total_Products_Sold".
total_product_sold = combined_data.groupby(['region', 'state', 'city'])[['units_sold']].sum()
total_product_sold = total_product_sold.rename(columns= {'units_sold':  'Total_Products_Sold'})

# Show Top 5 results
total_product_sold.sort_values('Total_Products_Sold', ascending=False).head()




```

#### Using `pivot_table`

```python
# Show the number products sold for region, state, and city.
pivoted_total_units = pd.pivot_table(combined_data,
                               values='units_sold',
                               index=['region', 'state',  'city'],
                               aggfunc='sum')




# Rename the "units_sold" column to "Total_Products_Sold"
pivoted_total_units = pivoted_total_units.rename(columns= {'units_sold':  'Total_Products_Sold'})

# Show the top 5 results.
pivoted_total_units.sort_values('Total_Products_Sold', ascending=False).head()
```

### 3. Determine which Region had the Most Sales

### 5. Determine which Retailer Sold the Most Women's Athletic Footwear

```python
# Filter the sales data to get the women's athletic footwear sales data.
filtered_data = combined_data[combined_data['product'] == "Women's Athletic Footwear"]

```
